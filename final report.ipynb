# Comparison of the Top 500 Words Between 2008 and 2018


```python
!pip install wordcloud
!pip install Pillow
from PIL import Image
import wordcloud
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import pandas
df = pandas.read_csv('2008.csv')
df.columns=['keyword','frequency']
df.set_index('keyword')
# filter top 500 words
word = df['keyword'][:500]

filepath = 'stopwords-zh.txt'
stopwords = [line.strip() for line in open(filepath, 'r', encoding='utf-8').readlines()]
def tag_cloud(text):
    mask = np.array(Image.open('china outline.png')) #set mask, you can change to the picture you like, but it must have a high color contrast
    wc = wordcloud.WordCloud(mode='RGBA',background_color='white',font_path='/Library/Fonts/Songti.ttc',max_words=2000,stopwords=stopwords,max_font_size=300,random_state=42,mask=mask)
    wc.generate_from_text(' '.join(text))
    plt.figure(figsize=(12,12))
    plt.imshow(wc, interpolation='bilinear')
    plt.axis("off")
    plt.title('2008 Top 500', loc='Center', fontsize=30)
    plt.show()
    return plt.show()
tag_cloud(word)
```

# Hot Words in 2008

The year of 2008 has undoubtedly been a year of transformation for China. Opportunities, crises and disasters have combined to create this special year. Since we cannot get the text of the first half of 2008, we only start to analyze it in the second half of 2008.

```python
import numpy as np
import matplotlib.pyplot as plt  
x1=['July','August','September','October','November', 'December']
y1=[1435, 2702, 661, 267, 117, 78]


l1=plt.plot(x1,y1,'g--')


plt.plot(x1,y1,'g+-')
plt.title('Olympic Games')
plt.xlabel('Month')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

The successful hosting of the 29th Olympic Games in Beijing on August 8, 2008 is a window for the world to truly understand and understand China. Therefore, from the text of July, it is easy to find that the country attaches great importance to the Olympic Games. The number of reports peaked in August and then declined successively.


```python
import numpy as np
import matplotlib.pyplot as plt  

x2=['July','August','September','October','November', 'December']
y2=[4,1,34,300,561,508]

l2=plt.plot(x2,y2,'r--')

plt.plot(x2,y2,'ro-')
plt.title('Financial Crisis')
plt.xlabel('Month')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

In the second half of the year, China experienced another major event: the financial crisis. From the frequency of words, the report of financial crisis starts from September and reaches its peak in November. This is because the crisis starts to get out of control in September.  In order to face the impact of the financial crisis, the government announced a major adjustment of fiscal and monetary policies in November. Based on the propaganda of national policies, People's Daily gives much more reports.

```python
import numpy as np
import matplotlib.pyplot as plt  
x1=['July','August','September','October','November', 'December']
y1=[0, 0, 192, 98, 2, 9]
x2=['July','August','September','October','November', 'December']
y2=[1,0,34,74,12,8]

l1=plt.plot(x1,y1,'g--',label='Milk Powder')
l2=plt.plot(x2,y2,'r--',label='Melamine')

plt.plot(x1,y1,'g+-', x2,y2,'ro-')
plt.title(' Melamine with Chinese Milk Scandal')
plt.xlabel('Month')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

Melamine with Chinese milk scandal is undoubtedly a tragedy in 2008. On September 11, the "Sanlu Milk Powder Incident" was exposed, which had a great impact and quickly attracted great attention from the central government. As we can be seen from the line chart, the word frequency of the "milk powder" reached the highest in September, when relevant events were reported. As people became aware of melamine, the investigation and notification published by the government began to appear in October. In December, the first trial of the case was sentenced, and the case again received media attention.

# Hot Words in 2018

According to Kesheng Ming, an associate professor of Fordham law school, the three most important words related with Xi's policy should be "One Belt One Road", "the Greater Bay" and "Xi's Socialism with Chinese Characteristics". 

```python
import pandas as pd
pd.read_csv('1718.csv')
df = pd.read_csv('1718.csv', header=None)
df.columns = ['word', 'frequency','month']
c = df['month']
d = [ str(i) for i in c ]
import matplotlib.pyplot as plt  

plt.plot(d,df['frequency'])
plt.title("One Belt One Road",fontsize=20)
plt.xlabel('month')
plt.ylabel("frequency",fontsize=14)
plt.xticks(rotation=45) 
plt.show()
```

"One Belt One Road" was most frequently mentioned in 2017, May with more than 2000 times, in the past year ended November, 30. That's because "One Belt One Road" forum was held in Beijing in May, 14 and 15, with presidents or prime ministers from 24 countries participating.

"Greater Bay Area" was firstly mentioned in Feb. 2017 and appeared frequently since the begining of 2018. That is because the relationship between mainland and Hong Kong became more and more intimate. In  2018, July, President Xi witnessed signing of the Framework Agreement for Deepening the Cooperation between Guangdong, Hong Kong and Macao to promote the construction of the Greater Bay. In addition, on 2018, September, 23, mainland-Hong Kong high-speed rail opened. 


```python
import pandas as pd
pd.read_csv('.csv')
df = pd.read_csv('.csv', header=None)
df.columns = ['word', 'frequency','month']
c = df['month']
d = [ str(i) for i in c ]
import matplotlib.pyplot as plt  

plt.plot(d,df['frequency'])
plt.title("Xi's Socialism with Chinese Characteristics",fontsize=20)
plt.xlabel('month')
plt.ylabel("frequency",fontsize=14)
plt.xticks(rotation=45) 
plt.show()
```

"Xi's Socialism with Chinese Characteristics" has become a common word in People's Daily in the last twelve months. Its word frequency exceeded 400 and 500 in 2017, October and 2018, March respectively. This is related with two historical decisons. 
On 2017, October, Xi's Socialism with Chinese Characteristics was added into constitution of Chinese Communist Party during the closing ceremony of 19th CPC National Congress. Following that, the 13th National Peoples Congress passed a Constitutional Amendment and decided to write it into People's Republic of China Constitution. After the two events, "Xi's Socialism with Chinese Characteristics" has maintained its high frequency in People's Daily.

The fact that the policy has gone through a transformation under Xis administration could be in inferred.

From People's daily, enormous changes have taken place in China.
