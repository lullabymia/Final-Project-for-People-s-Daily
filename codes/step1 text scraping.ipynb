```python
def gettext(url):
    import requests
    r=requests.get(url)
    r.encoding = 'utf-8'
    import bs4
    mypage = bs4.BeautifulSoup(r.text)
    text = mypage.find(attrs={'class': 'text_c'}).text.strip()
    return text 
urls = []
date = []
for c in range(1,32):
    if c < 10:
        m = str(c)
        c = '0'+ m
    else:
        c = c
    date.append(c)
date
for c in date:
    a = 1
    while a < 10:
        for b in range(1,5):
            url = "http://paper.people.com.cn/rmrb/html/2008-10/{2}/nw.D110000renmrb_200810{2}_{0}-0{1}.htm".format(a,b,c)
            urls.append(url)
           
        a = a + 1

```

```python
tendays=[]
for url in urls:
    try:
        text = gettext(url)
        tendays.append(text)
    except Exception as e:
        print(e)
        print('Error on page %s' % url)
```

```python


import jieba
words = open('200810月.text',"r").read()
words = jieba.cut(words)
filepath = 'stopwords-zh.txt'
stopwords = [line.strip() for line in open(filepath, 'r', encoding='utf-8').readlines()]
#remove stopwords
processed_word_list = []
for word in words:
    if word not in stopwords and len(word)>1: #remove single word
        processed_word_list.append(word)
wordcount={}
for word in processed_word_list:
    if word not in wordcount:
        wordcount[word] = 1
    else:
        wordcount[word] += 1
import operator
sorted_wordcount = sorted(wordcount.items(), key=operator.itemgetter(1),reverse=True)
```
